## Importing Libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Analysis

```python
data = pd.read_csv("ds_salaries.csv")

data.head()
```

```python
data.info()
```

```python
data.shape # Give the number of rows and column 
```

```python
data.describe()
```

```python
data.isnull().sum() #Checking null value present or not
```

```python
data[data.duplicated()]
```

```python
data.drop_duplicates() # Droping duplicates
```

```python
data['remote_ratio'].unique()
```

```python
def impact(X):
    if X == 0:
        return "On_site"
    elif X == 50:
        return "Hybrid"
    else:
        return "Remote"
    
```

```python
data = data.assign(Work_Environment=data['remote_ratio'].apply(impact))
```

## Data Visualization

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data['experience_level'],y=data['salary_in_usd'],hue=data['work_year'],errorbar=None)
plt.title("Salary based on Experience")
plt.xlabel("Experience Level")
plt.ylabel("Salary in USD")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data['work_year'],y=data['salary_in_usd'],hue=data['experience_level'],errorbar=None)
plt.title("Salaries over the Year")
plt.xlabel("Work Year")
plt.ylabel("Salary in USD")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data['experience_level'],y=data['salary'],hue=data['work_year'],errorbar=None)
plt.title("Salary based on Experience")
plt.xlabel("Experience Level")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(data['experience_level'].value_counts(),labels=data['experience_level'].unique(),autopct='%1.2f%%',shadow=True)
plt.title("Experience Level in data")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data['company_size'],y=data['salary_in_usd'],hue=data['work_year'],errorbar=None)
plt.title("Salary based on the size of the company over the Years")
plt.xlabel("Company Size")
plt.ylabel("Salary in USD")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data['company_size'],y=data['salary_in_usd'],hue=data['experience_level'],errorbar=None)
plt.title("Salary based on Company Size")
plt.xlabel("Company Size")
plt.ylabel("Salary in USD")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(data['Work_Environment'].value_counts(),labels=data['Work_Environment'].unique(),autopct='%1.2f%%',shadow=True)
plt.title("Work Environment")
plt.show()
```
