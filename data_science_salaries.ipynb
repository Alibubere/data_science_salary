## Data Visualization

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data['experience_level'],y=data['salary_in_usd'],hue=data['work_year'])
plt.title("Salary based on Experience")
plt.xlabel("Experience Level")
plt.ylabel("Salary in USD")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data['work_year'],y=data['salary_in_usd'],hue=data['experience_level'])
plt.title("Salaries over the Year")
plt.xlabel("Work Year")
plt.ylabel("Salary in USD")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(data['experience_level'].value_counts(),labels=data['experience_level'].unique(),autopct='%1.2f%%',shadow=True)
plt.title("Experience Level in data")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data['experience_level'],y=data['salary'],hue=data['work_year'])
plt.title("Salary based on Experience")
plt.xlabel("Experience Level")
plt.ylabel("Salary")
plt.show()
```

```python
data['company_size'].unique()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data['company_size'],y=data['salary_in_usd'],hue=data['work_year'])
plt.title("Salary based on the size of the company over the Years")
plt.xlabel("Company Size")
plt.ylabel("Salary in USD")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data['company_size'],y=data['salary_in_usd'],hue=data['experience_level'])
plt.title("Salary based on Company Size")
plt.xlabel("Company Size")
plt.ylabel("Salary in USD")
plt.show()
```
